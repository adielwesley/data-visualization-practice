```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```
# plot line
pontuacao = np.array([16, 13, 15, 12, 10, 20, 25, 40, 30, 25, 60, 50])
plt.plot(pontuacao)
plt.show()
```

```
# line color
plt.plot(pontuacao, c='r')
plt.show()
```

```
# line style
plt.plot(pontuacao, c='black', ls='--')
plt.show()
```

```
# markers
plt.plot(pontuacao, c='red', ls='--', marker='s')
plt.show()
```

```
# marker size
plt.plot(pontuacao, c='red', ls='--', marker='s', ms=18)
plt.show()
```

```
# two lines
pontuacao2 = np.array([33, 30, 40, 35, 30, 25, 10, 60, 45, 40, 50, 30])
plt.plot(pontuacao, c='red', ls='--', marker='s')
plt.plot(pontuacao2, c='blue', ls='--', marker='s')
plt.show()
```

```
# legend
plt.plot(pontuacao, c='red', ls='--', marker='s', label='primeira')
plt.plot(pontuacao2, c='blue', marker='s', label='segunda')
plt.legend()
plt.show()
```

```
# legend location
plt.plot(pontuacao, c='red', ls='--', marker='s', label='primeira')
plt.plot(pontuacao2, c='blue', marker='s', label='segunda')
plt.legend(loc='lower right')
plt.show()
```

```
# marker styles
plt.plot(pontuacao, c='red', ls='--', marker='s', label='primeira')
plt.plot(pontuacao2, c='blue', marker='^', label='segunda')
plt.legend(loc='lower right')
plt.show()
```

```
# three lines
pontuacao3 = np.array([39,42,37,37,40,35,28,45,45,50,55,40])
plt.plot(pontuacao, c='red', ls='--', marker='s', label='primeira')
plt.plot(pontuacao2, c='blue', marker='^', label='segunda')
plt.plot(pontuacao3, c='black', marker='s', label='terceira')
plt.legend(loc='lower right')
plt.show()
```

```
# title
plt.plot(pontuacao, c='red', ls='--', marker='s', label='primeira')
plt.plot(pontuacao2, c='blue', marker='^', label='segunda')
plt.plot(pontuacao3, c='black', marker='s', label='terceira')
plt.legend(loc='lower right')
plt.title('Avaliação de desempenho de modelos em 2019')
plt.show()
```

```
# axis legends
plt.plot(pontuacao, c='red', ls='--', marker='s', label='primeira')
plt.plot(pontuacao2, c='blue', marker='^', label='segunda')
plt.plot(pontuacao3, c='black', marker='s', label='terceira')
plt.legend(loc='lower right')
plt.title('Avaliação de desempenho de modelos em 2019')
plt.xlabel('Meses')
plt.ylabel('Acurácia')
plt.show()
```

```
# showing months at x axis
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun',
         'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']

plt.plot(meses, pontuacao, c='red', ls='--', marker='s', label='primeira')
plt.plot(meses, pontuacao2, c='blue', marker='^', label='segunda')
plt.plot(meses, pontuacao3, c='black', marker='s', label='terceira')
plt.legend(loc='lower right')
plt.title('Avaliação de desempenho de modelos em 2019')
plt.ylabel('Acurácia')
plt.show()
```

```
# changing line legends to Classifiers labels
plt.plot(meses, pontuacao, c='red', ls='--', marker='s', label='Decision Tree')
plt.plot(meses, pontuacao2, c='blue', marker='^', label='K-NN')
plt.plot(meses, pontuacao3, c='black', marker='s', label='Naive Bayes')
plt.legend(loc='lower right')
plt.title('Avaliação de desempenho de modelos em 2019')
plt.ylabel('Acurácia')
plt.show()
```

```
# improving visuals with seaborn
sns.set_theme()

plt.plot(meses, pontuacao, c='red', ls='--', marker='s', label='Decision Tree')
plt.plot(meses, pontuacao2, c='blue', marker='^', label='K-NN')
plt.plot(meses, pontuacao3, c='black', marker='s', label='Naive Bayes')
plt.legend(loc='lower right')
plt.title('Avaliação de desempenho de modelos em 2019')
plt.ylabel('Acurácia')
plt.show()
```

```
# fill areas between lines and x axis

plt.fill_between(meses, pontuacao, label='Decision Tree')
plt.fill_between(meses, pontuacao2, label='K-NN')
plt.fill_between(meses, pontuacao3, label='Naive Bayes')
plt.ylabel('Acurácia')
plt.legend(loc='upper left')
plt.show()
```

```
# applying transparency in areas between lines and x axis

plt.fill_between(meses, pontuacao, alpha=0.4, label='Decision Tree')
plt.fill_between(meses, pontuacao2, alpha=0.4, label='K-NN')
plt.fill_between(meses, pontuacao3, alpha=0.4, label='Naive Bayes')
plt.ylabel('Acurácia')
plt.legend(loc='upper left')
plt.show()
```

```
# showing upper contours of areas between lines and x axis

plt.fill_between(meses, pontuacao, alpha=0.4, label='Decision Tree')
plt.fill_between(meses, pontuacao2, alpha=0.4, label='K-NN')
plt.fill_between(meses, pontuacao3, alpha=0.4, label='Naive Bayes')
plt.ylabel('Acurácia')
plt.legend(loc='upper left')
plt.plot(meses, pontuacao)
plt.plot(meses, pontuacao2)
plt.plot(meses, pontuacao3)
plt.show()
```

```
# change style to stacked areas

models = ['Decision Tree', 'K-NN', 'Naive Bayes']
x=meses
y=[pontuacao, pontuacao2, pontuacao3]

# Plot
plt.margins(0,0)
plt.stackplot(x,y,labels=models,alpha=0.7)
plt.legend(loc='upper left')
plt.title('Gráfico de Áreas Empilhadas')
plt.show()
```

```
# change style to 100% stacked areas

# percentage calc for each model
total = pontuacao+pontuacao2+pontuacao3
perc1 = pontuacao/total
perc2 = pontuacao2/total
perc3 = pontuacao3/total

y = [perc1, perc2, perc3]

#plot
plt.stackplot(x,y,labels=models,alpha=0.7)
plt.margins(0,0)
plt.legend(loc='upper left')
plt.title('Gráfico de Áreas Empilhadas')
plt.show()
```

```
# circular area chart
from math import pi

# periods amount
models = ['Decision Tree', 'K-NN', 'Naive Bayes']
categories = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
N = len(categories)

# how many angles from each graphic axis
angles = [n / float(N) * 2 * pi for n in range(N)]
# closing the circle
angles += angles[:1]

# initializing the graphic
ax = plt.subplot(111, polar=True)

# To put the first axis on top
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)

# Defining markers and labels for x axis
plt.xticks(angles[:-1], categories)

# Defining y axis
ax.set_rlabel_position(0)
plt.yticks([15, 30, 45, 60], color="grey", size=10)
plt.ylim(0, 70)

# 'Decision Tree'
values = pontuacao.tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, label=models[0])
ax.fill(angles, values, 'b', alpha=0.1)

# 'K-NN'
values = pontuacao2.tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, label=models[1])
ax.fill(angles, values, 'r', alpha=0.1)

# 'Naive Bayes'
values = pontuacao3.tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, label=models[2])
ax.fill(angles, values, alpha=0.1)

# add legend
plt.legend(bbox_to_anchor=(0.1, 0.1))
plt.show()
```
